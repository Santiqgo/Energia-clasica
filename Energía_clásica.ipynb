# Energía clásica

Dada la energía cinética relativista de una partícula y su masa se hallara su energía cinética clásica.

$T_{clásica} = \frac{1}{2}M_{0}C^{2}\left(1-\left(\frac{M_{0}C^{2}}{T_{relativista} +M_{0}C^{2}}\right)^{2}\right)$

Donde si la masa en reposo $M_0$ de la partícula se entrega en $\frac{Mev}{C^2}$ entonces todas las $C^2$ desaparecen de la ecuación.

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def Tc(a,b): #Se define la función energía cinética clásica.
  return 0.5*a*(1-(a/(b+a))**2)
```

```python
Tc(0.511,1) #Ejemplo hecho en clase para observar que se entrega en resultado correcto.
```

Corriendo la siguiente linea se puede calcular la energía cinética clásica de una partícula conociendo su energía cinética relativista y masa en reposo.

```python
"""M_0=float(input("Introduce la masa en reposo: ")) # Masa en reposo de la partícula
T_r=float(input("Introduce la energía cinética relativista : ")) # Energía cinética relativista de la partícula

T_c=Tc(M_0,T_r)

print ("La energía cinética clásica de la particula es: %.5f Mev" %(T_c))
"""
```

```python
m_alfa=3728.44 #Mev/c^2
m_beta=0.511003 #Mev/c^2
m_neutron=939.5731 #Mev/c^2

T_r_alfa=np.linspace(2,10,10)
T_r_beta=np.linspace(1,5,10)
T_r_neutron=np.linspace(1,4,10)

T_c_alfa=Tc(m_alfa,T_r_alfa)
T_c_beta=Tc(m_beta,T_r_beta)
T_c_neutron=Tc(m_neutron,T_r_neutron)


plt.figure(figsize=(10 , 10))


plt.subplot(3,1,1)
plt.grid()
plt.xlabel("Energia relativista de la particula alfa")
plt.ylabel("Energia clásica de la particula alfa")
plt.plot(T_r_alfa,T_c_alfa,"-o")


plt.subplot(3,1,2)

plt.grid()
plt.xlabel("Energia relativista de la particula beta")
plt.ylabel("Energia clásica de la particula beta")
plt.plot(T_r_beta,T_c_beta,"-o")

plt.subplot(3,1,3)

plt.grid()
plt.xlabel("Energia relativista del neutron")
plt.ylabel("Energia clásica neutron")
plt.plot(T_r_neutron,T_c_neutron,"-o")

plt.show()
```

De las gráficas anteriores se puede observar que para partículas $\alpha$ y neutrones la relación entre la energía relativista y clásica es una linea recta cosa que no pasa para las particulas $\beta$.

Tomando el promedio entre todas las $\frac{T_{relativista}}{T_{clásica}}$.

```python
np.mean(T_r_alfa/T_c_alfa)
```

```python
np.mean(T_r_beta/T_c_beta)
```

```python
np.mean(T_r_neutron/T_c_neutron)
```

Lo que se puede ver en principio es como a mayor masa en reposo tiene una partícula, su energía cinética relativista se acercara a su energía cinética clásica, lo observado en los promedios tomados y de una forma simple en las figuras presentada.
